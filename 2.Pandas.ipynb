## Series 
A pandas 1d labeled array that can hold any data type, like single column in spreadsheet

```python
data = [100,101,102,104]
series = pd.Series(data) #Remember to use capital S not s for pd.Series()
print(series)
```

```python
data = [100.1,101.2,102,104]
series = pd.Series(data) #Remember to use capital S not s for pd.Series()
print(series)
```

```python
# you can have your own indexing
data = [100.1,101.2,102,104]
series = pd.Series(data, index=['a','b','c','d']) 
print(series)
#loc = location by label
print(series.loc['b']) # returns value at a
series.loc['b'] = 201.2
print(series.loc['b'])
```

```python
#iloc = integer location which is similar to list indexing in python
print(series.iloc[1])
```

```python
print(series[series>=102])
```

```python
# We can use dictionaries 
calories = {"Day1":1750,"Day2":2100,"Day3":1700}
series = pd.Series(calories)
print(series)
```

```python
series.loc["Day3"] += 500
print(series.loc['Day3'])
```

```python
print(series[series>=2000])
```

## DataFrames
A tabular data structure with rows and colums (2D). Similar to excel spreadsheet

```python
data = {"Name":['Venkat','Tarun','Kireety'],
        "Age":[20,22,24]}
df = pd.DataFrame(data)
print(df)
```

```python
#Custom indexing
data = {"Name":['Venkat','Tarun','Kireety'],
        "Age":[20,22,24]}
df = pd.DataFrame(data,index=['s1','s2','s3'])
print(df)
```

```python
print(df.loc["s2"])
print(df.iloc[0])
```

```python
#Add a new column
df["Job"] = ["CEO","HR","CTO"]

#Add a new row 
new_row = pd.DataFrame([{"Name":"Rehan","Age":22,"Job":"Chairman"}],
                       index=["s4"])
df = pd.concat([df,new_row]) #we create a new_row and concatnate to old df
print(df)
```

```python
#adding multiple rows
new_rows = pd.DataFrame([{"Name":"Varun","Age":26,"Job":"Lead"},
                         {"Name":"Arun","Age":29,"Job":"Manager"}],
                       index=["s5","s6"])
df = pd.concat([df,new_rows])
print(df)
```

## Filtering

```python
tall_pokemon = df[df["Height"] >=2 ]
print(tall_pokemon)
```

```python
legendary_pokemon = df[df["Legendary"]==1]
print(legendary_pokemon)
```

## Aggregate Functions 
Reduces a set of values into single summary value. Used to summarize and analyze data often used with groupby() function

```python
# finding mean for a column
print(df.mean(numeric_only=True)) # pandas cannot fing mean if not numeric so we are sying find mean of numeric columns
```

```python
print(df.sum(numeric_only=True))
```

```python
print(df.min(numeric_only=True))
print(df.max(numeric_only=True))
```

```python
print(df.count())
```

```python
# you can do those aggregate functions on selected columns
print(df["Height"].mean())
print(df["Weight"].median())
```

```python
# Group by -- grouping based on a particular property
group = df.groupby("Type1")
print(group) #this is a object when you print it if just give the memory location
             # but it returns a grouped series to work with
```

```python
print(group["Height"].mean())
```

```python
print(group["Weight"].mean())
```

```python
print(group["Weight"].min()) #minimum weight for each type
```

```python
print(group["Weight"].count()) 
```

## Data Cleaning 
The process of fixing/removing: incomplete,incorrect or irrelevant data. Pandas is mostly used for this

```python
df = pd.read_csv('sample.csv')
```

```python
# 1. Drop irrelevant colmns
df1 = df.drop(columns=["Legendary","No"])
print(df1)
```

```python
#2.Handle missing data
df = df.dropna(subset=["Type2"])
print(df)
```

```python
# Filling NaN values with other values
df = df.fillna({"Type2":"None"})
print(df)
```

```python
#3.Fix inconsistent values
df["Type1"] = df["Type1"].replace({"Grass": "GRASS", #replacing all type1 grass with GRASS
                                   "Fire":"FIRE"})
print(df)
```

```python
#4.Standardize text
df["Name"] = df["Name"].str.lower()
print(df)
```

```python
#5. Fix data types
df["Legendary"] = df["Legendary"].astype(bool)
print(df)
```

```python
#6.remove duplicate values
df = df.drop_duplicates()
print(df)
```
