```python
x = np.array([2023,2024,2025,2026]) 
y1 = np.array([15,25,30,20])
y2 = np.array([18,28,15,34])
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
# To customize the 2 line we can copy and paste but code becomes redundant
line_style = dict(marker="o",
             markersize=20,
             markerfacecolor="#31d69a",
             linestyle="solid",
             linewidth=4,
             color="#0a0a0a")
plt.plot(x,y1,**line_style) # Here we are just unpacking the dictinary using **
plt.plot(x,y2,**line_style) 
plt.show()
```

```python
line_style = dict(marker="o",
             markersize=20,
             markerfacecolor="#f020aa",
             linestyle="solid",
             linewidth=4,)
plt.plot(x,y1,color="#1236c7",**line_style) #If you want different colors you need to pass as keyword arguments
plt.plot(x,y2,color="#fc630b",**line_style) 
plt.show()
```

## Grid lines
Helps make plots easier to read by reference lines

```python
a = [1,2,3,4,5]
b = [5,10,15,10,5]
plt.grid() # if you dont pass any arguments it shows grid lines for whole chart 

plt.plot(a,b)
plt.show()
```

```python
a = [1,2,3,4,5]
b = [5,10,15,10,5]
plt.title("X axis Grid lines")
plt.grid(axis="x") # if you dont pass any arguments it shows grid lines for whole chart 

plt.plot(a,b)
plt.show()
plt.title("Y axis Grid lines")
plt.grid(axis="y",
         linewidth=3,
         color="lightgray",
         linestyle="dashed") # if you dont pass any arguments it shows grid lines for whole chart 

plt.plot(a,b)
plt.show()
```

## Bar Charts
Compare categories of data by representing each category with a bar

```python
food = ["Grains","Fruit","Vegetables","Protein","Dairy","Sweets"]
values = np.array([5,3,2,6,5,1])
plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")

plt.bar(food,values,color="black")
plt.show()
```

```python
# Horizontal bar chart
plt.barh(food,values,color="red")
plt.title("Daily Consumption")
plt.show()
```

## Pie Chart
Circular chart divided into slices to show percentages of the total
Good for visualizing distribution among categories

```python
categories = ["Freshmen","Sophomores","Juniors","Seniors"]
values = np.array([300,250,275,225])

plt.pie(values,labels=categories)
plt.show()
```

```python
#To get the percentage for each section
plt.pie(values,labels=categories,
                autopct="%1.1f%%") #autopct means auto percentage 1.1f limits digits 
plt.show()
```

```python
# You can give particular colours as a list
colors = ["red","yellow","blue","green"]
plt.title("Cal State")
plt.pie(values,labels=categories,
                autopct="%1.1f%%",
                colors=colors) #autopct means auto percentage 1.1f limits digits 
plt.show()
```

```python
plt.pie(values,labels=categories,
                autopct="%1.1f%%",
                explode=[0,0,0,0.2]) #Explode is used to move that particular partition outside
plt.show()
```

```python
#You can rotate the piechart
plt.pie(values,labels=categories,
                autopct="%1.1f%%",
                explode=[0,0,0,0.2],
                startangle=180) 
plt.show()
```

## Scatter plots 
Shows the relationship between two variables. Helps to identify a correlation(+,-,None)
Example: Study hours vs Test scores

```python
x = [0,1,1,2,3,4,5,6,7,7,8]
y = [55,60,65,62,68,70,75,78,82,85,87]
plt.xlabel("Hours studied")
plt.ylabel("Marks scored")
plt.title("Test Scores")
plt.scatter(x,y,color="skyblue",
                s=50) # Size of the dots
plt.show()
```

```python
x = [0,1,1,2,3,4,5,6,7,7,8]
y = [55,60,65,62,68,70,75,78,82,85,87]

a=np.array([0,1,2,2,3,4,5,6,7,8,8])
b=np.array([50,58,65,70,72,78,83,88,92,95,97])
plt.xlabel("Hours studied")
plt.ylabel("Marks scored")
plt.title("Test Scores")
plt.scatter(x,y,color="blue",label="Class A") #labeling different set of dots
plt.scatter(a,b,color="red",label="Class B")

plt.legend() #->Labels doesnt show up directly you need to write this inorder to see
plt.show()
```

## Histogram 
A visual representation of the distribution of quantitative data. They group values into bins(intervals) and counts how many fall in each range

```python
scores = np.random.normal(loc=80,scale=10,size=100) #loc means the middle part of normal distribution
                                                    # scale is standard deviation and size in no.of entries

plt.hist(scores)
plt.show()
```

```python
scores = np.clip(scores,0,100) #We are setting upper and lower bound so the scores doesnt go over 100 or below 0
plt.hist(scores)
plt.show()
```

```python
plt.hist(scores,bins=8,   #In total we will be having 8 bins
         color="green",edgecolor="black") 
plt.show()
```

## Subplots 
1. Figure = The entire canvas
2. Ax = A single plot(subplot). A fingure can contains many Ax's

```python
print(plt.subplots(2,2)) 
# this return a tuple with figure and array of axes acording to given number
# so what we will do is we unpack them and then do plt.show()
```

```python
figure,axes = plt.subplots(3,3)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
figure , axes = plt.subplots(2,2)
axes[0,0].plot(x, x+2,color="red")
axes[0,0].set_title("x+2")
axes[0,1].plot(x, x/2,color="blue")
axes[0,1].set_title("x/2")
axes[1,0].plot(x, x**3,color="green")
axes[1,0].set_title("x**3")
axes[1,1].plot(x, x**2,color="yellow")
axes[1,1].set_title("x**2")
plt.show()
```

```python
# To avoid overlapping
x = np.array([1,2,3,4,5])
figure , axes = plt.subplots(2,2)
axes[0,0].plot(x, x+2,color="red")
axes[0,0].set_title("x+2")
axes[0,1].plot(x, x/2,color="blue")
axes[0,1].set_title("x/2")
axes[1,0].plot(x, x**3,color="green")
axes[1,0].set_title("x**3")
axes[1,1].plot(x, x**2,color="yellow")
axes[1,1].set_title("x**2")
plt.tight_layout() # This will make sure that the graphs do not overlap
plt.show()
```

## Pandas+Matplotlib

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sample.csv')
print(df)
```

```python
type_count = df["Type1"].value_counts()
plt.barh(type_count.index,type_count.values)
plt.show()
```

```python
type_count = df["Type1"].value_counts(ascending=True) #large to small
plt.barh(type_count.index,type_count.values,color='yellow',
                                            edgecolor='black')
plt.title("Counts of pokemons")
plt.xlabel("Counts")
plt.ylabel("Type")
plt.show()
```
