# 📔 Understanding Logging in Python

**Logging** is like a "black box" flight recorder for your code. It records events, errors, and status updates while your program runs so you don't have to guess what happened if things go wrong.

---

## 🚀 Why Logging Beats `print()`
While `print()` is okay for tiny scripts, it fails in professional projects because:
1. **Urgency Levels:** You can distinguish between a "minor note" and a "system crash."
2. **Persistence:** Logs go to **files** that stay after the program closes; `print()` disappears.
3. **Easy Control:** You can turn off "Debug" messages with one line of code instead of deleting 50 `print()` statements.
4. **Context:** Logs automatically include **Time, File Name, and Line Number.**

---

## 📊 The 5 Logging Levels
Python uses these levels to categorize importance. By default, Python only shows **WARNING** and above.

| Level | Value | Purpose |
| :--- | :--- | :--- |
| **DEBUG** | 10 | Detailed info for diagnosing problems (Variable values, etc.). |
| **INFO** | 20 | Confirmation that things are working as expected. |
| **WARNING** | 30 | Something unexpected happened, but the app is still running. |
| **ERROR** | 40 | A serious problem—the software couldn't perform a specific task. |
| **CRITICAL** | 50 | A total failure—the program might shut down. |

## 🛠️ Implementation Code
*Copy this into a **Code Cell** in your notebook:*

```python
import logging

# Configure the logger
logging.basicConfig(
    level=logging.DEBUG, 
    format='%(asctime)s - %(levelname)s - %(message)s'
)

# Practice using the levels
logging.info("Starting the SQLite import process...")
logging.debug("Variable x is currently set to 100.")
logging.warning("Disk space is getting low (85% full).")
logging.error("Failed to delete the 'employees' table.")
logging.critical("Database connection lost! Shutting down.")
```

````python
import logging

## Configure the basic logging settings
logging.basicConfig(level=logging.DEBUG)  # Set the lowest level to capture

## Log messages with different severity levels
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
````

````python
## Configuring more in logging
import logging                                   #Restart kernel before running this because you already set the the logging config in previous cell
logging.basicConfig(
    level=logging.DEBUG,
    format = '%(asctime)s-%(name)s-%(levelname)s-%(message)s', 
    datefmt = '%Y-%m-%d %H:%M:%S'
)

logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
````

````python
import logging                                   #Restart kernel before running this because you already set the the logging config in previous cell
logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format = '%(asctime)s-%(name)s-%(levelname)s-%(message)s', 
    datefmt = '%Y-%m-%d %H:%M:%S'
)

logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message") # Now you can see that all log messages will be saved to that app.log file 
````
