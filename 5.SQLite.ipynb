```python
#sqlite is default package
import sqlite3
```

```python
## Connect to an SQLite database
connection = sqlite3.connect('example.db') #Creates a example.db file and establishes a connection
connection                                 # You can work with that database locally
```

We need to create a cursor then we can use it iterate line by line in db

```python
cursor = connection.cursor() 
```

Creating a table

```python
cursor.execute('''
Create Table If Not Exists employees(
               id Integer Primary Key,
               name Text Not Null,
               age Integer,
               department text)
''')
## We need to commit the changes
connection.commit()
```

```python
cursor.execute('''
Select * from employees
''')
```

```python
## Inserting the data in sqlite table 
cursor.execute('''
Insert Into employees(name,age,department)
               values('Krish','32','Data Scientist')
''')

cursor.execute('''
Insert Into employees(name,age,department)
               values('Krish','40','CTO')
''')

cursor.execute('''
Insert Into employees(name,age,department)
               values('Tarun','22','Data Analyst')
''')

connection.commit()
```

```python
## Query the data from table
cursor.execute('''
Select * from employees
''')
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
cursor.execute('''
    SELECT * FROM employees 
    LIMIT 2
''')                            # The idea is you run the sql commands and then use cursor object to print 
connection.commit()
for row in cursor:
    print(row)
```

```python
## Update the data in the table
cursor.execute('''
Update employees Set age=34 
               Where name='Krish'
''')
```

```python
cursor.execute('''
Select * from employees
''')
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
cursor.execute('DROP TABLE IF EXISTS employees')

# Commit the change to the database
connection.commit()
```
