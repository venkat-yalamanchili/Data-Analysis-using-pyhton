```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('housing.csv')
print(df)
```

```python
g = sns.relplot(kind='scatter',data=df,x='longitude',y='latitude')
       # type of plot i.e ralational then kind,dataframe,x and y axis
g.set(title="Geographic location")
```

```python
g = sns.relplot(kind='scatter',data=df,x='longitude',y='latitude',hue='median_house_value')
g.set(title="Geographic location")
```

```python
g = sns.relplot(kind='scatter',data=df,x='longitude',y='latitude',hue='ocean_proximity')
g.set(title="Geographic location")
```

```python
# We add a new column above_median with binary values 1 or 0
df['above_median_house_value'] = (df.median_house_value > df.median_house_value.median()).astype(int)
print(df)
```

```python
g = sns.relplot(kind='scatter',data=df,x='longitude',y='latitude',hue='above_median_house_value')
g.set(title="Geographic location")
```

```python
g = sns.relplot(kind='scatter',data=df,x='longitude',y='latitude',
                style='above_median_house_value')
```

```python
## Basic plotting with seaborn
tips = sns.load_dataset('tips') ##Using inbuit data sets
tips
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter plot of bill vs tip")
plt.show()
```

```python
## Line plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.show()
```

```python
## Categorical plots
## Bar plots
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.barplot(x='smoker',y='total_bill',data=tips)
```

```python
## Box plot
sns.boxplot(x='day',y='total_bill',data=tips) #In the box plot we use a metric called percentile
```

```python
## Violine Plot
sns.violinplot(x='day',y='total_bill',data=tips) # Middle white dash is the mean of that day
```

```python
## Histograms 
sns.histplot(tips['total_bill'],bins=10)
```

```python
# You can also get the shape of the hostogram using kde = kernel density
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
## You can just plot the shape i.e, kernel density
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
## Pair plot
sns.pairplot(tips)
```

```python
## Heat Map - coreleation between values
## Remember cannot used with categorial features
corr = tips[['total_bill','tip','size']].corr() 
print(corr)
```

The above table is the co relation between 3 variables i.e., if total_bill is changes by 1 unit tip will be changed by 0.6 in +ve way

```python
sns.heatmap(corr)
```

```python
sns.heatmap(corr,annot=True,cmap='cool') #annot means annotations i.e,values inside the box
```
